```python
# https://matplotlib.org/stable/gallery/shapes_and_collections/compound_path.html#sphx-glr-gallery-shapes-and-collections-compound-path-py
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# version 1 making image with numpy and displaying it with matplotlib
# maaking image
img = np.zeros(
    (500, 500, 3), np.uint8
)  # 500,500 size of image, 255 is color filling, 3 is the color channel, uint8 is type
# image color filling
img.fill(255)
# 0 to 255, 255 is white 0 is black

# showing the image
plt.imshow(img)
```

```python
# drawing on image
rectangle = cv2.rectangle(img, (100, 100), (300, 200), (0, 255, 0), thickness=4)
# first(x,y) (100,100) is the starting point, second (300,200) is the ending point, third (0,255,0) is the color, fourth thickness=4 is the thickness of the rectangle
plt.imshow(img);
```

```python
# changing the thickness to -1 to see part III 1.2
rectangle = cv2.rectangle(img, (100, 100), (300, 200), (0, 255, 0), thickness=-1)
```

```python
# display the image
plt.imshow(img);

# why does it fill the rectangle?
# https://docs.opencv.org/4.x/dc/da5/tutorial_py_drawing_functions.html
# According to the opencv documentation, the thickness=-1 will fill the shapes easily
# without adding extra parameters
```

```python
# now we have a shape as well as an image, lets put texts on the rectangle that is in image
# set the text
text = "Sangjune"
# set the font
font = cv2.FONT_HERSHEY_SIMPLEX
# set the font size
font_size = 1
# set the color
color = (0, 0, 255)
# set the thickness
thickness = 2
# put the text on the image
rectangle = cv2.putText(img, text, (100, 150), font, font_size, color, thickness)
# first is the image, second is the text, third is the starting point, fourth is the font, fifth is the font size, sixth is the color, seventh is the thickness, eighth is the line type
# display the image
plt.imshow(img);
```

```python
# version 2 lets load an image and do the same as above
img = cv2.imread("Lucy2.jpg")
plt.imshow(img);  # seems like the image is in BGR format
```

```python
# turn it into RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

# now we have an image, lets put a rectangle on it
rectangle = cv2.rectangle(img, (100, 100), (300, 200), (0, 255, 0), thickness=4)
# display the image
plt.imshow(img);

#
```

```python
# change thickness -1
rectangle = cv2.rectangle(img, (100, 100), (300, 200), (0, 255, 0), thickness=-1)
# display the image
plt.imshow(img);
```

```python
# lets put a text in the rectangle
text = "Sangjune"
# set the font
font = cv2.FONT_HERSHEY_SIMPLEX
# set the font size
font_size = 1
# set the color
color = (0, 0, 255)
# set the thickness
thickness = 2
# put the text on the image
rectangle = cv2.putText(img, text, (100, 150), font, font_size, color, thickness)
# display the image
plt.imshow(img);
```

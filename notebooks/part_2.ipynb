```python
#importing the necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

#reading the image
img1 = cv2.imread('Lucy.jpg')

#BGR to RGB change in plt
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

# displaying image1
# https://matplotlib.org/stable/tutorials/images.html
plt.figure(figsize=(15,5))
plt.imshow(img1)
plt.title('Original Image 1')
plt.axis('off')



```

```python
#brightness control
# lets put the orginal image and the brightness controlled image side by side

# Display original image
plt.subplot(131)
plt.imshow(img1)
plt.title('Original')
plt.axis('off')

# Increase brightness by adding 150 to all pixels
#https://numpy.org/doc/stable/user/basics.broadcasting.html
# we are adding 50 to all the pixels of the image as (RGB)
bright_img = cv2.add(img1, np.array([150, 150, 150]))
plt.subplot(132)
plt.imshow(bright_img)
plt.title('Brightness (+50)')
plt.axis('off')
```

```python
#contrast control
# Display original image
plt.subplot(131)
plt.imshow(img1)
plt.title('Original')
plt.axis('off')

# Increase brightness by adding 150 to all pixels
#https://numpy.org/doc/stable/user/basics.broadcasting.html
# we are adding 50 to all the pixels of the image as (RGB)
bright_img = cv2.add(img1, np.array([150, 150, 150]))
plt.subplot(132)
plt.imshow(bright_img)
plt.title('Brightness (+50)')
plt.axis('off')
# adding contracts by multiplying the image by 0,5
contrast_img = cv2.multiply(img1, np.array([0.5, 0.5, 0.5]))
plt.subplot(133)
plt.imshow(contrast_img)
plt.title('Contrast (*0.5)')
plt.axis('off')
```

```python
# second part of part 2
# Load the second image

img2 = cv2.imread('Dog and Cat.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)


# check the sizing
print("Image 1 dimensions:", img1.shape)
print("Image 2 dimensions:", img2.shape)

# img 2 is bigger lets make sure it matches with img1 's sizing
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
print("New image 2 dimensions:", img2.shape)
# https://www.geeksforgeeks.org/image-resizing-using-opencv-python/

# Display both original images side by side
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img1)
plt.title('Image 1')
plt.axis('off')

plt.subplot(122)
plt.imshow(img2)
plt.title('Image 2')
plt.axis('off')
plt.show()
```

```python
# setting alpha value from user input
# https://www.w3schools.com/python/ref_func_input.asp
# make suer to accept float
alpha = float(input("Enter blend factor (0-1): "))

# implementing linear blending
blended = cv2.addWeighted(img1, 1-alpha, img2, alpha, 0)
# https://docs.opencv.org/3.4/d5/dc4/tutorial_adding_images.html
#blended =  (1 - alpha) * img1 + alpha * img2

# Display the blended image
plt.figure(figsize=(10,8))
plt.imshow(blended)
plt.title(f'Blended Image (alpha={alpha})')
plt.axis('off')
plt.show()
```
